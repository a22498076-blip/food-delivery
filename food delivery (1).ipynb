```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("food_delivery_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
total_orders = len(df)
print("Total Orders:", total_orders)
```

```python
popular_food = df['food_item'].value_counts()

print(popular_food.head(10))
```

```python
top_food = df['food_item'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_food.plot(kind='bar')

plt.title("Top 10 Food Items")
plt.xlabel("Food Item")
plt.ylabel("Number of Orders")

plt.show()
```

```python
top_locations = df['location'].value_counts()

print(top_locations.head(10))
```

```python
location_data = df['location'].value_counts().head(5)

plt.figure(figsize=(7,7))
location_data.plot(kind='pie', autopct='%1.1f%%')

plt.title("Orders by Location")
plt.ylabel("")

plt.show()
```

```python
avg_order = df['order_value'].mean()

print("Average Order Value:", avg_order)
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['order_value'], bins=20)

plt.title("Order Value Distribution")
plt.xlabel("Order Value")
plt.ylabel("Frequency")

plt.show()
```

```python
delivery_method = df['delivery_method'].value_counts()

print(delivery_method)
```

```python
delivery_method.plot(kind='bar')

plt.title("Delivery Methods")
plt.xlabel("Method")
plt.ylabel("Count")

plt.show()
```

```python
avg_satisfaction = df['customer_satisfaction'].mean()

print("Average Customer Satisfaction:", avg_satisfaction)
```

```python
df['customer_satisfaction'].value_counts().sort_index().plot(kind='bar')

plt.title("Customer Satisfaction Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Customers")

plt.show()
```
